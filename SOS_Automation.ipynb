```python
# function for taking out the $  sign if it is a string value

def clean_currency(x):
    
    if isinstance(x, str):
        return(x.replace('$', ''))
    return(x)
```

```python
# applying function and changing data type to float

R1_df['Sale Price'] = R1_df['Sale Price'].apply(clean_currency).astype('float')
R1_df
```

```python
# Checking datatype after applying function

R1_df["Sale Price"].apply(type).value_counts()
```

```python
# taking out $20 filter

testall2 = R1_df[R1_df['Sale Price'] >= 20.00]
testall2
```

```python
# dropping row values in receive date last that do not have a receive date

testall3 = testall2.dropna(subset=['Receive Date Last'])
testall3
```
